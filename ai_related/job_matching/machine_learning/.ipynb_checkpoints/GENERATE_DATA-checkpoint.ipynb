# Data Generation — Job Matching ML

Generates synthetic but realistic data mirroring production schema
(see `backend/livedb.txt`).

## Feature set (18)

| Group | Features |
|---|---|
| Semantic (2) | `cosine_sim` (freelancer source-text × job source-text), `portfolio_relevance` (max recency-weighted past-work cosine: manual portfolio ∪ completed contracts) |
| Skill fit (5) | `skill_overlap_pct`, `skill_required_matched`, `skill_required_total`, `skill_preferred_pct`, `skill_depth` (proficiency-weighted) |
| Experience fit (3) | `experience_level_match`, `exp_delta`, `recency_score` (half-life 24mo) |
| Budget fit (2) | `rate_in_budget`, `rate_ratio` |
| Context (3) | `language_match`, `speciality_match`, `domain_match` |
| Profile depth (3) | `has_portfolio`, `work_exp_count`, `total_jobs` |

**Why no `bio_job_cosine`?** The freelancer source-text used by `cosine_sim` already
includes the bio + skills + work experience, so a separate bio×job cosine would be a
noisier subset of the same signal — it would split feature importance between
correlated features without adding new information.

**Why `portfolio_relevance` is kept:** it answers a different question — *"have you
done this kind of work before?"* `cosine_sim` averages everything into one vector;
`portfolio_relevance` takes the max over individual past projects + completed
contracts, so a freelancer who has built ETL pipelines specifically scores higher
than one whose profile *generally* mentions data engineering.

## Embeddings — Ollama `nomic-embed-text` (768-dim)

Same model production uses (see `backend/.env: OLLAMA_TEXT_EMBEDDING`). 01 runs in
your VPS docker container, so `host.docker.internal:11434` reaches the host Ollama.
Embeddings are cached to `data/embeddings_cache.parquet` — first run takes ~3 min,
re-runs are instant.

## Label — non-linear with feature interactions

```
fit_score = (
  0.22 × (skill_overlap × (0.5 + 0.5 × exp_match))      # skill × exp
+ 0.18 × (skill_depth × √max(recency, 0.05))            # depth × recency
+ 0.14 × (cosine_sim × skill_overlap)                   # cosine × skill
+ 0.10 × √portfolio_relevance     [if has past work]    # past evidence
+ 0.08 × domain_fit
+ 0.08 × cosine_sim
+ 0.06 × rate_in_budget
+ 0.06 × language_match
+ 0.04 × speciality_match
+ 0.04 × profile_depth
) × over_qualification_multiplier                       # non-monotonic in exp_delta

fit_score += N(0, 0.04)                                 # unobserved noise
P(label=1) = sigmoid(9 × (fit_score − 0.50))            # ~30% positive rate
```

**Cold-start fairness:** the `√portfolio_relevance` term is only added when the
freelancer has past-work evidence (work_experience.count > 0 or any portfolio
entries). A new freelancer with strong skills isn't penalised for absence of
past work.

**Output:** CSV files in `data/` — one per entity + `ml_training_pairs.csv`,
plus `embeddings_cache.parquet` (re-used across runs).


````python
import numpy as np
import pandas as pd
import uuid
import os
import time
import json
import hashlib
import random
import requests
from datetime import date, timedelta

np.random.seed(42)
random.seed(42)

DATA_DIR = 'data'
os.makedirs(DATA_DIR, exist_ok=True)
os.makedirs('models', exist_ok=True)

# Ollama embedding endpoint — same model production uses (see backend/.env).
# 01 runs inside the VPS docker container, so host.docker.internal resolves
# to the host where Ollama is running.
OLLAMA_URL   = os.getenv('OLLAMA_URL', 'http://host.docker.internal:11434/api/generate')
OLLAMA_MODEL = os.getenv('OLLAMA_TEXT_EMBEDDING', 'nomic-embed-text')
OLLAMA_EMBED_URL = OLLAMA_URL.split('/api/')[0].rstrip('/') + '/api/embeddings'

EMBED_CACHE_PATH = os.path.join(DATA_DIR, 'embeddings_cache.parquet')
EMBED_DIM        = 768

print(f'Ollama:        {OLLAMA_EMBED_URL} | model={OLLAMA_MODEL}')
print(f'Embed cache:   {EMBED_CACHE_PATH}')
print('Setup complete')
````

````python
# ── Embedding helper (Ollama) with disk cache ─────────────────────────────────
# Caches by sha256(text) so re-running 01 doesn't re-embed unchanged strings.
# Each text → 768-dim vector. Falls back to a deterministic hash-based vector
# if Ollama isn't reachable, so the notebook still runs end-to-end (with a
# warning that Ollama-quality semantics are missing).

if os.path.exists(EMBED_CACHE_PATH):
    _embed_df = pd.read_parquet(EMBED_CACHE_PATH)
    _embed_cache: dict[str, list[float]] = {
        row['hash']: list(row['vector']) for _, row in _embed_df.iterrows()
    }
    print(f'Loaded {len(_embed_cache):,} cached embeddings from {EMBED_CACHE_PATH}')
else:
    _embed_cache = {}
    print('No embedding cache yet — building from scratch')

_OLLAMA_OK: bool | None = None

def _ollama_reachable() -> bool:
    global _OLLAMA_OK
    if _OLLAMA_OK is not None:
        return _OLLAMA_OK
    try:
        r = requests.post(OLLAMA_EMBED_URL,
                          json={'model': OLLAMA_MODEL, 'prompt': 'ping'},
                          timeout=8)
        _OLLAMA_OK = r.status_code == 200 and 'embedding' in r.json()
    except Exception as e:
        print(f'WARN: Ollama not reachable ({e}) — using deterministic fallback')
        _OLLAMA_OK = False
    return _OLLAMA_OK

def _hash_fallback(text: str) -> list[float]:
    """Deterministic 768-dim pseudo-embedding from text hash. Same text → same vector.
    Useless for semantics but keeps the notebook runnable when Ollama is down."""
    rng = np.random.default_rng(int(hashlib.sha256(text.encode()).hexdigest()[:16], 16))
    v = rng.standard_normal(EMBED_DIM)
    return (v / np.linalg.norm(v)).tolist()

def embed(text: str) -> list[float]:
    """Return a 768-dim embedding for `text`, cached by sha256(text)."""
    if not text or not text.strip():
        return [0.0] * EMBED_DIM
    h = hashlib.sha256(text.encode('utf-8')).hexdigest()
    if h in _embed_cache:
        return _embed_cache[h]
    if _ollama_reachable():
        try:
            r = requests.post(OLLAMA_EMBED_URL,
                              json={'model': OLLAMA_MODEL, 'prompt': text},
                              timeout=60)
            r.raise_for_status()
            vec = r.json().get('embedding', [])
            if not vec:
                print(f"DEBUG: Ollama returned empty vector for hash {h[:8]}... — falling back")
                vec = _hash_fallback(text)
        except Exception:
            print(f"DEBUG: Ollama request failed ({e}) for hash {h[:8]}... — falling back")
            vec = _hash_fallback(text)
    else:
        print(f"DEBUG: Ollama request failed ({e}) for hash {h[:8]}... — falling back")
        vec = _hash_fallback(text)
    _embed_cache[h] = vec
    return vec

def embed_batch(texts: list[str], desc: str = 'embed') -> list[list[float]]:
    """Embed a list of strings sequentially; reports progress every 50."""
    out = []
    t0 = time.perf_counter()
    for i, t in enumerate(texts):
        out.append(embed(t))
        if (i + 1) % 50 == 0:
            print(f'  {desc}: {i+1}/{len(texts)} ({(time.perf_counter()-t0):.1f}s)')
    print(f'  {desc}: done {len(texts)} in {(time.perf_counter()-t0):.1f}s '
          f'({len(_embed_cache)} unique cached)')
    return out

def cosine(a: list[float], b: list[float]) -> float:
    """Cosine similarity between two vectors. Returns 0 if either is zero-norm."""
    av, bv = np.asarray(a), np.asarray(b)
    na, nb = np.linalg.norm(av), np.linalg.norm(bv)
    if na == 0 or nb == 0:
        return 0.0
    return float(np.dot(av, bv) / (na * nb))

def save_embedding_cache():
    rows = [{'hash': h, 'vector': v} for h, v in _embed_cache.items()]
    pd.DataFrame(rows).to_parquet(EMBED_CACHE_PATH, index=False)
    print(f'Saved {len(rows):,} embeddings → {EMBED_CACHE_PATH}')

# Probe Ollama once so the user sees the status now, not 100 jobs in
print(f'Ollama reachable: {_ollama_reachable()}')
````

````python
# ── Freelancers (mirrors production: freelancer table) ────────────────────────
# Production schema: freelancer_id, user_id, full_name, bio,
#   estimated_rate, rate_time, rate_currency, total_jobs
N_FREELANCERS = 700

FIRST_NAMES = ['Alice','Bob','Charlie','Diana','Edward','Fatima','George','Hannah',
               'Ivan','Julia','Kevin','Lily','Marco','Nina','Oscar','Priya',
               'Quinn','Rachel','Samuel','Tanya','Umar','Victoria','William','Xena',
               'Yusuf','Zara','Ahmad','Budi','Citra','Dewi','Eko','Faisal','Gita',
               'Hendra','Indah','Joko','Kartika','Luthfi','Maya','Naufal','Putri']
LAST_NAMES  = ['Smith','Johnson','Williams','Brown','Jones','Garcia','Miller','Davis',
               'Santoso','Wijaya','Hidayat','Kusuma','Sari','Pratama','Utomo',
               'Lee','Kim','Chen','Wang','Kumar','Patel','Singh','Ahmed','Nakamura']

RATE_RANGES = {
    'backend':(20,80),'frontend':(15,70),'fullstack':(25,90),
    'mobile':(20,75),'ml_ai':(35,120),'design':(15,60),
    'devops':(30,100),'blockchain':(40,150),
}
RATE_TIMES = ['hourly', 'hourly', 'hourly', 'daily', 'weekly', 'monthly', 'annually']

APPROX_CURRENCY_TO_USD = {
    'USD': 1.0,
    'IDR': 16000.0,
    'EUR': 0.93,
}

# ── Diversified vocabulary for richer bios (Path B) ─────────────────────────
# Each domain has ~30 phrases (was 10), grouped by sub-area so we can sample
# from different sub-areas within one bio. This gives nomic-embed-text genuinely
# different vectors for two freelancers in the same domain — a Backend dev who
# focuses on databases will embed differently from one who focuses on APIs.
DOMAIN_BIO_PHRASES = {
    'backend': {
        'apis':       ['REST API design', 'GraphQL gateway services', 'gRPC microservice contracts',
                       'API versioning strategies', 'webhook delivery with retries',
                       'API rate limiting and quotas', 'OpenAPI documentation'],
        'data':       ['PostgreSQL schema design', 'database query optimisation',
                       'Redis caching layers', 'event sourcing pipelines',
                       'CQRS read-model projections', 'time-series storage with TimescaleDB',
                       'data partitioning at scale'],
        'platform':   ['Docker container orchestration', 'background job queues with Celery',
                       'async I/O patterns in Python', 'multi-tenant SaaS architectures',
                       'ETL pipelines for analytics', 'observability stacks (OpenTelemetry)'],
        'security':   ['OAuth2 + OIDC integration', 'API security audits',
                       'rate limiting against abuse', 'secret rotation workflows'],
    },
    'frontend': {
        'ui':         ['responsive web interfaces', 'pixel-perfect Figma implementation',
                       'design system components', 'CSS architecture (BEM, ITCSS)',
                       'cross-browser compatibility'],
        'react':      ['React component libraries', 'React Server Components',
                       'state management with Zustand', 'TanStack Query for data fetching',
                       'custom hook patterns'],
        'perf':       ['performance optimisation', 'Core Web Vitals tuning',
                       'lazy loading and code splitting', 'animation with Framer Motion',
                       'bundle size budgets'],
        'a11y':       ['accessibility (WCAG 2.1 AA)', 'screen-reader testing',
                       'keyboard navigation', 'colour-contrast audits'],
    },
    'fullstack': {
        'product':    ['MVP prototyping for early-stage startups', 'B2B SaaS features end-to-end',
                       'subscription billing with Stripe', 'usage-based metering',
                       'feature flag rollouts'],
        'stack':      ['React + Node.js stacks', 'Next.js App Router projects',
                       'TypeScript across the stack', 'tRPC for end-to-end type safety',
                       'Prisma + PostgreSQL', 'Supabase backends'],
        'infra':      ['CI/CD on GitHub Actions', 'AWS deployment with Terraform',
                       'preview environments per PR', 'real-time features via WebSocket',
                       'serverless on Vercel'],
        'practices':  ['test-driven product development', 'incremental migration strategies',
                       'feature-branch workflows'],
    },
    'mobile': {
        'native':     ['native iOS in Swift/SwiftUI', 'native Android in Kotlin/Jetpack Compose',
                       'platform-specific UX patterns', 'in-app purchases and subscriptions'],
        'cross':      ['Flutter cross-platform delivery', 'React Native bridges',
                       'Expo managed workflow', 'shared business logic in Dart/TypeScript'],
        'ux':         ['offline-first sync patterns', 'push notifications with FCM',
                       'animation-rich interactions', 'deep linking and universal links',
                       'app store releases and beta tracks'],
        'perf':       ['battery profiling', 'cold-start optimisation',
                       'memory leak hunting in Instruments'],
    },
    'ml_ai': {
        'modelling':  ['production ML pipelines', 'transformer fine-tuning (LoRA, QLoRA)',
                       'gradient-boosted trees for tabular data', 'recommendation system retrieval',
                       'sequence-to-sequence models'],
        'nlp':        ['NLP preprocessing and tokenisation', 'named-entity recognition',
                       'document RAG with vector stores', 'prompt engineering for LLMs',
                       'function-calling agents'],
        'cv':         ['computer vision systems', 'object detection with YOLO',
                       'image segmentation with U-Net', 'OCR pipelines'],
        'ops':        ['feature engineering at scale', 'experiment tracking with MLflow',
                       'model serving APIs with FastAPI', 'GPU autoscaling on Kubernetes',
                       'data labelling workflows'],
    },
    'design': {
        'systems':    ['design systems and tokens', 'component libraries in Figma',
                       'multi-brand theming'],
        'ux':         ['user research and interviews', 'usability testing protocols',
                       'mobile and web UX flows', 'information architecture',
                       'cross-cultural UX considerations'],
        'visual':     ['high-fidelity Figma prototypes', 'visual identity and branding',
                       'micro-interactions and motion', 'illustration and iconography',
                       'data visualisation design'],
        'process':    ['design-to-developer handoff', 'accessibility audits (WCAG)',
                       'workshop facilitation', 'design ops tooling'],
    },
    'devops': {
        'orchestration': ['Kubernetes cluster operations', 'Helm chart authoring',
                          'service-mesh with Istio', 'pod autoscaling strategies'],
        'iac':           ['Terraform infrastructure-as-code', 'Pulumi for typed IaC',
                          'GitOps with ArgoCD', 'immutable infrastructure'],
        'observability': ['observability stacks (Prometheus/Grafana)', 'distributed tracing with Jaeger',
                          'log aggregation with Loki', 'on-call incident response',
                          'SLO design and error budgets'],
        'cost':          ['cost-optimisation on AWS/GCP', 'spot-instance workflows',
                          'right-sizing audits', 'reserved-instance planning'],
        'security':      ['secrets and IAM management', 'container hardening',
                          'CIS benchmark compliance'],
    },
    'blockchain': {
        'contracts':  ['Solidity smart contracts', 'Vyper alternative implementations',
                       'gas-efficient patterns', 'token standards (ERC-20/721/1155)',
                       'upgradeable proxy patterns'],
        'protocols':  ['DeFi protocol integration', 'AMM design and slippage handling',
                       'lending and borrowing primitives', 'bridge protocols'],
        'tooling':    ['Hardhat and Foundry workflows', 'subgraph indexing with The Graph',
                       'wallet-connect flows', 'multi-sig governance with Gnosis Safe'],
        'security':   ['on-chain audits', 'formal verification basics',
                       'fuzzing with Echidna', 'vulnerability disclosure workflows'],
    },
}

# Sentence templates — bio uses 3 to 6 sentences from this pool, randomly ordered
BIO_INTRO_TEMPLATES = [
    "{seniority} working primarily on {domain_label}.",
    "{seniority} with a focus on {domain_label} engagements.",
    "Based remotely, {seniority_lower} delivering {domain_label} work.",
    "{domain_label} contributor — {seniority_lower}.",
    "{seniority} who has spent the last {years} years on {domain_label}.",
]
BIO_SKILLS_TEMPLATES = [
    "Strong with {p1} and {p2}.",
    "Day-to-day work involves {p1}, {p2}, and {p3}.",
    "Recent engagements have leaned on {p1} alongside {p2}.",
    "Comfortable shipping {p1}; growing depth in {p2}.",
    "Most projects touch {p1} or {p2} at some point.",
    "Toolbox: {p1}, {p2}, {p3}, and increasingly {p4}.",
]
BIO_HISTORY_TEMPLATES = [
    "Track record covers {history}.",
    "{history} so far, mostly with early-stage product teams.",
    "Engagements to date: {history}.",
    "Background: {history}, ranging from MVPs to scale-ups.",
    "{history} — happy to share case studies on request.",
]
BIO_STYLE_TEMPLATES = [
    "Aim for maintainable code, frequent demos, and clear async communication.",
    "Prefer tight feedback loops and shipping in two-week iterations.",
    "Comfortable owning unclear scope and breaking it into milestones.",
    "Strong on documentation and on-boarding teammates.",
    "Bias toward writing tests early and keeping pull requests small.",
    "Used to working across timezones with weekly written updates.",
    "Care a lot about edge cases and observability from day one.",
]
BIO_LANG_TEMPLATES = [
    "Working in English and Bahasa Indonesia.",
    "Communicate fluently in English; native Bahasa Indonesia.",
    "Open to remote collaboration in English or Bahasa Indonesia.",
    "Comfortable with English-speaking standups and Indonesian client calls.",
    "",  # sometimes omit the language sentence entirely
]

def _seniority_words(total_jobs: int):
    if total_jobs == 0:
        return ("Fresh practitioner", "fresh practitioner", random.randint(0, 1))
    if total_jobs < 3:
        return ("Junior practitioner", "junior practitioner", random.randint(1, 2))
    if total_jobs < 10:
        return ("Mid-career engineer", "mid-career engineer", random.randint(3, 5))
    return (f"Senior specialist with {total_jobs}+ projects",
            f"senior specialist with {total_jobs}+ projects",
            random.randint(5, 12))

def _history_phrase(total_jobs: int) -> str:
    if total_jobs == 0:
        return "no commercial projects yet but two open-source side projects"
    if total_jobs < 3:
        return f"{total_jobs} small commercial project{'s' if total_jobs != 1 else ''}"
    if total_jobs < 10:
        return f"{total_jobs} client engagements across SaaS, marketplace, and internal-tool products"
    return f"{total_jobs}+ engagements across multiple industries"

def make_rich_bio(domain: str, total_jobs: int, name: str) -> str:
    """Generate a varied 3–6 sentence bio.

    Uses random sentence count, random sentence ordering (within reason), and
    samples vocabulary from sub-area buckets so two freelancers in the same
    domain produce genuinely different embedding vectors. This is what lets
    nomic-embed-text give meaningful cosine_sim signal in training."""
    sub_areas = list(DOMAIN_BIO_PHRASES[domain].keys())
    # Pick 2-4 sub-areas this freelancer is "known for" — gives intra-domain variety
    chosen_subs = random.sample(sub_areas, k=random.randint(2, min(4, len(sub_areas))))
    pool = []
    for sub in chosen_subs:
        pool.extend(DOMAIN_BIO_PHRASES[domain][sub])
    # Pick 4-6 phrases from the pool without replacement
    n_phrases = min(random.randint(4, 6), len(pool))
    phrases = random.sample(pool, n_phrases)
    p1, p2, p3, p4 = (phrases + ['related tooling'] * 4)[:4]

    domain_label = domain.replace('_', '/').title()
    seniority, seniority_lower, years = _seniority_words(total_jobs)
    history = _history_phrase(total_jobs)

    # Choose how many sentences (3 to 6) and sample without ordering bias
    sentences = []
    sentences.append(random.choice(BIO_INTRO_TEMPLATES).format(
        seniority=seniority, seniority_lower=seniority_lower,
        domain_label=domain_label, years=years))
    sentences.append(random.choice(BIO_SKILLS_TEMPLATES).format(
        p1=p1, p2=p2, p3=p3, p4=p4))
    if random.random() < 0.85:
        sentences.append(random.choice(BIO_HISTORY_TEMPLATES).format(history=history))
    if random.random() < 0.75:
        sentences.append(random.choice(BIO_STYLE_TEMPLATES))
    lang_sentence = random.choice(BIO_LANG_TEMPLATES)
    if lang_sentence:
        sentences.append(lang_sentence)
    # Optional extra freeform phrase mention from a different sub-area for cross-domain noise
    if random.random() < 0.30 and len(sub_areas) > len(chosen_subs):
        other_sub = random.choice([s for s in sub_areas if s not in chosen_subs])
        extra_phrase = random.choice(DOMAIN_BIO_PHRASES[domain][other_sub])
        sentences.append(f"Side interest: {extra_phrase}.")

    return " ".join(sentences)


freelancers = []
for i in range(N_FREELANCERS):
    domain   = random.choice(ALL_DOMAINS)
    level_roll = random.random()
    if level_roll < 0.25:
        total_jobs = random.randint(0, 2)
    elif level_roll < 0.75:
        total_jobs = random.randint(3, 9)
    else:
        total_jobs = random.randint(10, 45)

    rmin, rmax   = RATE_RANGES[domain]
    rate_time    = random.choice(RATE_TIMES)
    currency     = random.choices(['USD', 'IDR', 'EUR'], weights=[5, 4, 1])[0]
    rate_usd_equiv   = random.uniform(rmin, rmax)
    estimated_rate   = round(rate_usd_equiv * APPROX_CURRENCY_TO_USD.get(currency, 1.0), 2)
    full_name = f'{random.choice(FIRST_NAMES)} {random.choice(LAST_NAMES)}'

    freelancers.append({
        'freelancer_id':  str(uuid.uuid4()),
        'user_id':        str(uuid.uuid4()),
        'full_name':      full_name,
        'bio':            make_rich_bio(domain, total_jobs, full_name),
        'estimated_rate': estimated_rate,
        'rate_time':      rate_time,
        'rate_currency':  currency,
        'total_jobs':     total_jobs,
        'domain':         domain,
    })

freelancers_df = pd.DataFrame(freelancers)
new_count  = (freelancers_df['total_jobs'] <= 2).sum()
mid_count  = ((freelancers_df['total_jobs'] >= 3) & (freelancers_df['total_jobs'] <= 9)).sum()
expert_cnt = (freelancers_df['total_jobs'] >= 10).sum()
currency_counts = freelancers_df['rate_currency'].value_counts().to_dict()

# Sample 3 bios from the SAME domain so we can eyeball the diversification.
sample_domain = freelancers_df['domain'].iloc[0]
same_domain_bios = freelancers_df[freelancers_df['domain'] == sample_domain].head(3)
print(f'Freelancers: {len(freelancers_df)}')
print(f'  entry (0-2 jobs):    {new_count}')
print(f'  intermediate (3-9):  {mid_count}')
print(f'  expert (10+):        {expert_cnt}')
print(f'  currencies:          {currency_counts}')
print(f'\nSample bios from same domain ({sample_domain}) — should look DIFFERENT:')
for _, r in same_domain_bios.iterrows():
    print(f'  • [{len(r["bio"].split())} words] {r["bio"][:200]}...')
````

````python
# ── Freelancer attributes ─────────────────────────────────────────────────────
# Mirrors: freelancer_skill, freelancer_speciality, freelancer_language,
#          work_experience, education, portfolio
fl_skills_rows, fl_specs_rows, fl_langs_rows = [], [], []
work_exp_rows, education_rows, portfolio_rows = [], [], []

PROF_SKILL = ['expert','advanced','intermediate','beginner']
PROF_W     = [0.20, 0.40, 0.30, 0.10]

# Map proficiency name → numeric weight for skill_depth feature.
# MUST match _PROFICIENCY_WEIGHT in ml_ranker.py.
PROFICIENCY_WEIGHT = {
    'expert':       1.00,
    'advanced':     0.75,
    'intermediate': 0.50,
    'beginner':     0.30,
}

# Domain-flavoured project blurbs for richer portfolio embeddings.
# Same idea as bios: enough variation that nomic-embed-text gives signal.
DOMAIN_PROJECT_BLURBS = {
    'backend':    ['ETL pipeline ingesting nightly batches into PostgreSQL with Airflow scheduling',
                   'OAuth2 + JWT auth service handling 5k requests per minute',
                   'Stripe-style webhook receiver with idempotent retries and dead-letter handling',
                   'GraphQL gateway aggregating three internal services into one API surface',
                   'Background worker queue using Redis streams for invoice generation'],
    'frontend':   ['Customer dashboard with charting, filtering, and CSV export — handles 200k rows',
                   'Marketing landing page with WebGL hero and 95+ Lighthouse score',
                   'Multi-step onboarding flow with autosave and validation across 12 fields',
                   'Real-time chat UI with typing indicators, read receipts, and file uploads',
                   'Admin panel with role-based access control and granular permission editing'],
    'fullstack':  ['MVP for a fitness tracking SaaS — Next.js app, PostgreSQL, Stripe billing',
                   'B2B vendor portal with invoice generation, e-signing, and audit trail',
                   'Booking platform integrating Google Calendar, video calls, and SMS reminders',
                   'Inventory management web app with barcode scanning on mobile and desktop',
                   'Subscription analytics dashboard pulling from MRR, churn, and LTV metrics'],
    'mobile':     ['Food-delivery rider app with GPS tracking, push notifications, in-app payments',
                   'Banking transaction history with biometric login and offline view',
                   'Health-tracking iOS app with HealthKit and Apple Watch companion',
                   'Marketplace Android app supporting saved searches and chat negotiation',
                   'Travel itinerary builder with airline schedule sync and weather overlays'],
    'ml_ai':      ['Customer-support classifier reducing first-response time by 40%',
                   'Document RAG pipeline retrieving from 50k internal PDFs with citation tracking',
                   'Image-tagging model deployed via FastAPI with GPU autoscaling on Kubernetes',
                   'Sales-forecast model integrated into a finance dashboard, MAPE under 8%',
                   'Resume parser extracting structured fields with confidence scores'],
    'design':     ['Brand redesign for an EdTech startup — logo, type system, web and print collateral',
                   'Mobile app redesign cutting onboarding drop-off from 60% to 25%',
                   'Design system covering 80 components with light/dark themes and tokens',
                   'Marketing site for a B2B SaaS product including illustrations and motion',
                   'Internal tool UX overhaul reducing support tickets by 30%'],
    'devops':     ['Migrated a monolith to ECS Fargate, cutting deploy time from 25 to 4 minutes',
                   'Set up Terraform modules for repeatable VPC, RDS, and IAM provisioning',
                   'Implemented blue/green deploys with automated rollback on failed health checks',
                   'Centralised logging with Loki + Grafana, retention tuned for 30 days',
                   'On-call rotation with PagerDuty integration and runbook automation'],
    'blockchain': ['NFT minting contract with lazy-mint pattern saving 60% gas vs eager mint',
                   'DEX router supporting Uniswap V3 and SushiSwap with slippage protection',
                   'DAO governance module with weighted voting and on-chain proposal storage',
                   'Cross-chain bridge between Ethereum and Polygon using lock-and-mint',
                   'Audited and patched a lending protocol — disclosed two medium findings'],
}

JOB_TITLE_PROJECT_PHRASES = {
    'backend': ['backend engineer', 'API developer'],
    'frontend': ['frontend engineer', 'web developer'],
    'fullstack': ['full-stack engineer', 'web developer'],
    'mobile': ['mobile developer', 'app engineer'],
    'ml_ai': ['ML engineer', 'data scientist'],
    'design': ['designer', 'UX practitioner'],
    'devops': ['DevOps engineer', 'platform engineer'],
    'blockchain': ['smart contract developer', 'blockchain engineer'],
}

today = date.today()

for _, f in freelancers_df.iterrows():
    fid    = f['freelancer_id']
    domain = f['domain']
    total  = int(f['total_jobs'])

    # Skills: 3-9 primary domain skills + 1-2 cross-domain
    primary  = DOMAIN_SKILLS[domain]
    n_pick   = random.randint(3, min(9, len(primary)))
    chosen   = random.sample(primary, n_pick)
    other_sk = [s for d, sl in DOMAIN_SKILLS.items() if d != domain
                for s in sl if s not in chosen and s in skill_id_map]
    chosen  += random.sample(other_sk, min(random.randint(1, 2), len(other_sk)))
    seen = set()
    for sk in chosen:
        if sk in skill_id_map and sk not in seen:
            seen.add(sk)
            fl_skills_rows.append({
                'freelancer_skill_id': str(uuid.uuid4()),
                'freelancer_id': fid, 'skill_id': skill_id_map[sk],
                'skill_name': sk,
                'proficiency_level': random.choices(PROF_SKILL, weights=PROF_W)[0],
            })

    # Specialities
    for j, sp in enumerate(random.sample(DOMAIN_SPECIALITIES[domain],
                                          random.randint(1, min(3, len(DOMAIN_SPECIALITIES[domain]))))):
        fl_specs_rows.append({
            'freelancer_speciality_id': str(uuid.uuid4()),
            'freelancer_id': fid, 'speciality_id': spec_id_map[sp],
            'speciality_name': sp, 'is_primary': int(j == 0),
        })

    # Languages
    if random.random() < 0.65:
        chosen_langs = ['Indonesian']
        extra = random.sample([l for l in LANGUAGES if l != 'Indonesian'], random.randint(0, 2))
        chosen_langs += extra
    else:
        chosen_langs = random.sample(LANGUAGES, random.randint(1, 3))
    for j, lang in enumerate(chosen_langs):
        fl_langs_rows.append({
            'freelancer_language_id': str(uuid.uuid4()),
            'freelancer_id': fid, 'language_id': lang_id_map[lang],
            'language_name': lang,
            'proficiency_level': 'native' if j == 0 else random.choice(['basic','conversational','fluent']),
        })

    has_history = total > 0
    if has_history:
        # ── Work experience: most-recent-first, dated. Drives recency_score. ─
        n_exp = min(4, max(1, total // 4 + 1))
        end = today
        for j in range(random.randint(1, n_exp)):
            dur   = random.randint(6, 36)
            start = end - timedelta(days=dur * 30)
            domain_label = domain.replace('_',' ').title()
            description_blurb = random.choice(DOMAIN_PROJECT_BLURBS[domain])
            work_exp_rows.append({
                'work_experience_id': str(uuid.uuid4()), 'freelancer_id': fid,
                'job_title':    ('Senior ' if j == 0 else '') + domain_label + ' Engineer',
                'company_name': random.choice(['TechCorp','StartupXYZ','BigCo','Digital Agency',
                                               'Consultancy','Fintech Co','E-Commerce Ltd']),
                'start_date':   start.isoformat(),
                'end_date':     None if j == 0 else (end - timedelta(days=30)).isoformat(),
                'is_current':   int(j == 0),
                'description':  description_blurb,
            })
            end = start - timedelta(days=30)

        if random.random() < 0.55:
            education_rows.append({
                'education_id': str(uuid.uuid4()), 'freelancer_id': fid,
                'institution_name': random.choice(['State University','Tech Institute','UI','ITS','ITB','NUS']),
                'degree': random.choice(['B.Sc Computer Science','B.Eng Software Engineering',
                                         'B.Sc Information Technology','M.Sc Computer Science']),
                'field_of_study': domain.replace('_',' ').title(),
                'start_date': '2018-08-01', 'end_date': '2022-06-01', 'is_current': 0,
            })

        # ── Portfolio: mix of auto-generated (from contracts) and manual.
        # Auto rows mirror "completed contracts" — these will NOT be embedded in
        # portfolio_embedding (Path B). Manual rows are user-curated showcase items.
        # Each gets a richer description so embeddings carry signal.
        n_total_port = random.randint(0, 5)
        for k in range(n_total_port):
            # 60% chance a portfolio entry is auto (came from a completed contract).
            # In real production this matches: most freelancers have more contract history
            # than self-curated showcase items.
            is_auto = total > 0 and random.random() < 0.6
            blurb = random.choice(DOMAIN_PROJECT_BLURBS[domain])
            domain_label = domain.replace('_', ' ').title()
            project_age_months = random.randint(1, 48)
            completion_dt = today - timedelta(days=int(project_age_months * 30))
            if is_auto:
                title = f"{domain_label} Engagement — {completion_dt.year}"
                description = f"Completed project: {blurb}. Delivered for client engagement."
            else:
                title = f"{domain_label} Showcase: {blurb[:50]}"
                description = (
                    f"{blurb}. Personal/showcase project; full case-study available on request. "
                    f"Tech stack reflects core {domain_label} expertise."
                )
            portfolio_rows.append({
                'portfolio_id':         str(uuid.uuid4()),
                'freelancer_id':        fid,
                'project_title':        title,
                'project_description':  description,
                'completion_date':      completion_dt.isoformat(),
                'is_auto_generated':    int(is_auto),
                # contract_id stays None for synthetic data; production would link it.
            })

fl_skills_df   = pd.DataFrame(fl_skills_rows)
fl_specs_df    = pd.DataFrame(fl_specs_rows)
fl_langs_df    = pd.DataFrame(fl_langs_rows)
work_exp_df    = pd.DataFrame(work_exp_rows)
education_df   = pd.DataFrame(education_rows)
portfolio_df   = pd.DataFrame(portfolio_rows)

n_auto   = int(portfolio_df['is_auto_generated'].sum()) if len(portfolio_df) else 0
n_manual = len(portfolio_df) - n_auto
print(f'freelancer_skills:       {len(fl_skills_df)}')
print(f'freelancer_specialities: {len(fl_specs_df)}')
print(f'freelancer_languages:    {len(fl_langs_df)}')
print(f'work_experience:         {len(work_exp_df)}')
print(f'education:               {len(education_df)}')
print(f'portfolio:               {len(portfolio_df)} (auto={n_auto}, manual={n_manual})')
````

````python
# ── Job posts, roles, skills (mirrors production: job_post, job_role, job_role_skill) ─
N_JOBS = 600
BUDGET_RANGES = {
    'backend':(500,5000),'frontend':(300,4000),'fullstack':(800,8000),
    'mobile':(600,6000),'ml_ai':(1000,10000),'design':(300,3000),
    'devops':(800,8000),'blockchain':(1000,15000),
}

# Per-domain "concrete deliverable" pool — much larger than before, grouped by
# project type so two jobs in the same domain can describe very different work.
JOB_DELIVERABLE_POOLS = {
    'backend': [
        'Build a REST API for a stock-take mobile app, including OAuth2 login, pagination, and offline retry support',
        'Design a PostgreSQL schema for multi-tenant invoicing with row-level security',
        'Refactor a monolithic Django backend into 3 microservices behind an API gateway',
        'Implement a Stripe-style webhook receiver with idempotent retries and dead-letter queue',
        'Build a background job queue using Redis streams to send transactional emails',
        'Add full-text search to an existing FastAPI service using PostgreSQL tsvector',
        'Migrate a legacy MySQL database to PostgreSQL with zero downtime',
        'Design a rate-limiting layer for a public API serving 10k requests per minute',
        'Build an internal admin tool with audit log and granular permissions',
        'Implement event sourcing for an e-commerce order lifecycle',
    ],
    'frontend': [
        'Build a customer dashboard with charting, filtering, and CSV export — must handle 200k rows',
        'Create a marketing landing page with WebGL hero animation and 95+ Lighthouse score',
        'Implement a multi-step onboarding flow with autosave and validation across 12 fields',
        'Refactor an Angular app to React with feature parity over 4 weeks',
        'Build a real-time chat UI with typing indicators, read receipts, and file uploads',
        'Create a design-system library with 30 components, dark mode, and Storybook docs',
        'Add accessibility (WCAG 2.1 AA) compliance to an existing SaaS product',
        'Optimise an existing React app from 6s LCP to under 2s on 3G',
        'Build an interactive data-viz dashboard using D3 and React',
        'Implement an editable Kanban board with drag-and-drop and live multi-user updates',
    ],
    'fullstack': [
        'Build the MVP for a fitness-tracking SaaS — Next.js app, PostgreSQL, Stripe billing',
        'Create a B2B vendor portal with invoice generation, e-signing, and audit trail',
        'Develop a booking platform integrating Google Calendar, video calls, and SMS reminders',
        'Build an inventory management web app with barcode scanning on mobile and desktop',
        'Implement a subscription analytics dashboard pulling from MRR, churn, and LTV metrics',
        'Build a marketplace with two-sided onboarding, search, chat, and Stripe Connect payouts',
        'Develop a content management system with role-based publishing and versioning',
        'Create an internal billing tool with multi-currency invoicing and tax calculations',
        'Build a peer-to-peer rental platform with calendar availability and damage deposits',
    ],
    'mobile': [
        'Build a food-delivery rider app with GPS tracking, push notifications, and in-app payments',
        'Create a banking transaction-history app with biometric login and offline view',
        'Develop a health-tracking iOS app with HealthKit and an Apple Watch companion',
        'Build a marketplace Android app supporting saved searches and chat negotiation',
        'Implement a travel itinerary builder with airline schedule sync and weather overlays',
        'Migrate an existing Cordova app to Flutter with feature parity',
        'Add live video streaming to an existing app using WebRTC',
        'Build a habit-tracking app with streaks, reminders, and Apple/Google Health sync',
    ],
    'ml_ai': [
        'Build a customer-support classifier to route tickets — should reduce first-response time by 40%',
        'Implement a document RAG pipeline retrieving from 50k internal PDFs with citation tracking',
        'Train an image-tagging model and deploy via FastAPI with GPU autoscaling on Kubernetes',
        'Build a sales-forecast model integrated into a finance dashboard, MAPE under 8%',
        'Create a resume parser extracting structured fields with confidence scores',
        'Fine-tune an LLM on internal knowledge base for an internal Q&A bot',
        'Build a churn prediction model with explainability via SHAP for the success team',
        'Implement an anomaly detection system on telemetry data with weekly retraining',
        'Develop a recommendation system for a media platform — collaborative + content-based hybrid',
    ],
    'design': [
        'Brand redesign for an EdTech startup — logo, type system, web and print collateral',
        'Mobile app redesign — measurably reduce onboarding drop-off (currently 60%)',
        'Build a design system covering 80 components with light/dark themes and tokens',
        'Marketing site for a B2B SaaS product including illustrations and motion',
        'Internal tool UX overhaul to reduce support tickets by 30%',
        'User research study (12 interviews + synthesis) for a fintech product',
        'Design an iconography set (200 icons) following a custom grid system',
        'Create high-fidelity prototypes for a new feature flow with usability testing rounds',
    ],
    'devops': [
        'Migrate a monolith to ECS Fargate, cutting deploy time from 25 to 4 minutes',
        'Set up Terraform modules for repeatable VPC, RDS, and IAM provisioning',
        'Implement blue/green deploys with automated rollback on failed health checks',
        'Centralise logging with Loki + Grafana, retention tuned for 30 days',
        'Set up an on-call rotation with PagerDuty integration and runbook automation',
        'Build a self-service deploy platform for internal teams using GitHub Actions',
        'Audit AWS spend and implement right-sizing recommendations to cut 30% cost',
        'Design a multi-region active-active setup with global load balancing',
        'Implement secret rotation across services using AWS Secrets Manager',
    ],
    'blockchain': [
        'Develop an NFT minting contract with lazy-mint pattern saving 60% gas vs eager mint',
        'Build a DEX router supporting Uniswap V3 and SushiSwap with slippage protection',
        'Create a DAO governance module with weighted voting and on-chain proposal storage',
        'Implement a cross-chain bridge between Ethereum and Polygon using lock-and-mint',
        'Audit and patch a lending protocol — disclose findings with severity ratings',
        'Build a yield-aggregator vault with composable strategy modules',
        'Develop a wallet-connect onboarding flow for a Web3 game',
        'Implement a subgraph indexer for an existing protocol with 6 entity types',
    ],
}

# Sentence templates for job descriptions — many variants, randomly composed
JOB_INTRO_TEMPLATES = [
    "We are looking to hire a {exp_lvl}-level contributor for a remote {domain_label} engagement.",
    "Our team needs an {exp_lvl} {domain_label} freelancer to support an active project.",
    "Open contract: {domain_label} work suited to a {exp_lvl} contributor.",
    "Seeking a {domain_label} specialist (level: {exp_lvl}) for a 4–12 week engagement.",
    "Project posting: {domain_label} contractor at the {exp_lvl} level.",
]
JOB_SCOPE_TEMPLATES = [
    "The main scope is: {d1}.",
    "Primary deliverable: {d1}.",
    "We need someone to {d1_lower}.",
    "Core work: {d1}.",
]
JOB_SECONDARY_TEMPLATES = [
    "Secondary tasks may include {d2_lower}.",
    "There is also some adjacent work around {d2_lower}.",
    "On top of that, you may help with {d2_lower}.",
    "Stretch goal if time allows: {d2}.",
    "",  # sometimes only one deliverable
]
JOB_REQUIREMENTS_TEMPLATES = [
    "You should be comfortable owning the work end-to-end and writing brief weekly updates.",
    "Prior experience shipping similar projects is required — please link relevant work.",
    "Good written English is required for async stand-ups; Bahasa Indonesia is a plus.",
    "Familiarity with our broader stack (see job role skills) is expected.",
    "We expect clear documentation of decisions and trade-offs.",
    "Two-week iteration cadence with one async demo per cycle.",
]
JOB_LOGISTICS_TEMPLATES = [
    "Delivery is in two-week iterations with a demo at the end of each cycle.",
    "Engagement is fully remote; team timezones are GMT+7 and GMT−5.",
    "Payment is milestone-based; budget includes bug fixes within 30 days of delivery.",
    "Weekly written status updates expected, with one async video walkthrough per milestone.",
    "Code is reviewed in pull requests; mainline must always be deployable.",
    "",
]

def make_rich_job_description(domain: str, exp_lvl: str) -> str:
    """Generate a varied 3–6 sentence job description.

    Uses concrete project deliverables (10 per domain) and randomly composes
    sentence templates so two jobs in the same domain look genuinely different
    in their text — the embedding then gives a real signal for cosine_sim."""
    pool = JOB_DELIVERABLE_POOLS[domain]
    deliverables = random.sample(pool, k=2)
    d1, d2 = deliverables
    domain_label = domain.replace('_', '/').title()

    sentences = []
    sentences.append(random.choice(JOB_INTRO_TEMPLATES).format(exp_lvl=exp_lvl, domain_label=domain_label))
    sentences.append(random.choice(JOB_SCOPE_TEMPLATES).format(d1=d1, d1_lower=d1[0].lower() + d1[1:]))
    secondary = random.choice(JOB_SECONDARY_TEMPLATES)
    if secondary:
        sentences.append(secondary.format(d2=d2, d2_lower=d2[0].lower() + d2[1:]))
    if random.random() < 0.85:
        sentences.append(random.choice(JOB_REQUIREMENTS_TEMPLATES))
    logistics = random.choice(JOB_LOGISTICS_TEMPLATES)
    if logistics:
        sentences.append(logistics)
    return " ".join(sentences)


job_posts_rows, job_roles_rows, job_role_skills_rows = [], [], []

for _ in range(N_JOBS):
    domain   = random.choice(ALL_DOMAINS)
    jp_id    = str(uuid.uuid4())
    exp_lvl  = random.choices(['entry','intermediate','expert'], weights=[3,5,2])[0]
    bmin, bmax = BUDGET_RANGES[domain]

    title_template = random.choice(JOB_TITLE_TEMPLATES[domain])
    adj = random.choice(ADJECTIVES)
    job_title = title_template.format(adj=adj).strip()

    job_posts_rows.append({
        'job_post_id':      jp_id,
        'client_id':        str(uuid.uuid4()),
        'job_title':        job_title,
        'job_description':  make_rich_job_description(domain, exp_lvl),
        'project_type':     random.choice(['individual','team']),
        'project_scope':    random.choice(['small','medium','large']),
        'experience_level': exp_lvl,
        'status':           random.choices(['active','closed','filled'], weights=[6,2,2])[0],
        'domain':           domain,
    })

    for role_idx in range(random.randint(1, 3)):
        jr_id = str(uuid.uuid4())
        job_roles_rows.append({
            'job_role_id':  jr_id, 'job_post_id': jp_id,
            'role_title':   domain.replace('_',' ').title() + (' Designer' if domain == 'design' else ' Engineer'),
            'role_budget':  round(random.uniform(bmin, bmax), 2),
            'budget_currency': 'USD',
            'budget_type':  random.choice(['fixed','negotiable']),
            'display_order': role_idx,
        })

        dom_sk   = DOMAIN_SKILLS[domain]
        n_req    = min(random.randint(2, 5), len(dom_sk))
        req_sk   = random.sample(dom_sk, n_req)
        other_sk = [s for d, sl in DOMAIN_SKILLS.items()
                    if d != domain for s in sl if s not in req_sk]
        pref_sk  = random.sample(other_sk, min(random.randint(0, 2), len(other_sk)))

        seen = set()
        for sk in req_sk:
            if sk in skill_id_map and sk not in seen:
                seen.add(sk)
                job_role_skills_rows.append({
                    'job_role_skill_id': str(uuid.uuid4()),
                    'job_role_id': jr_id, 'job_post_id': jp_id,
                    'skill_id': skill_id_map[sk], 'skill_name': sk,
                    'is_required': True, 'importance_level': 'required',
                    'job_domain': domain, 'exp_level': exp_lvl,
                })
        for sk in pref_sk:
            if sk in skill_id_map and sk not in seen:
                seen.add(sk)
                job_role_skills_rows.append({
                    'job_role_skill_id': str(uuid.uuid4()),
                    'job_role_id': jr_id, 'job_post_id': jp_id,
                    'skill_id': skill_id_map[sk], 'skill_name': sk,
                    'is_required': False,
                    'importance_level': random.choice(['preferred','nice_to_have']),
                    'job_domain': domain, 'exp_level': exp_lvl,
                })

job_posts_df       = pd.DataFrame(job_posts_rows)
job_roles_df       = pd.DataFrame(job_roles_rows)
job_role_skills_df = pd.DataFrame(job_role_skills_rows)

sample_domain = job_posts_df['domain'].iloc[0]
same_domain_jobs = job_posts_df[job_posts_df['domain'] == sample_domain].head(3)
print(f'job_posts:      {len(job_posts_df)}')
print(f'job_roles:      {len(job_roles_df)}')
print(f'job_role_skills:{len(job_role_skills_df)}')
print(f'\nSample job descriptions from same domain ({sample_domain}) — should look DIFFERENT:')
for _, r in same_domain_jobs.iterrows():
    print(f'  • [{len(r["job_description"].split())} words] {r["job_description"][:200]}...')
````

````python
# ── Feature engineering helpers ───────────────────────────────────────────────
# Must produce the same values as ml_ranker.py _compute_job_features() (18 features).

import math

# Freelancer lookups
fl_skill_set     = fl_skills_df.groupby('freelancer_id')['skill_name'].apply(set).to_dict()
fl_skill_prof: dict[str, dict[str, str]] = {}
for _, r in fl_skills_df.iterrows():
    fl_skill_prof.setdefault(r['freelancer_id'], {})[r['skill_name']] = r['proficiency_level']

fl_lang_set      = fl_langs_df.groupby('freelancer_id')['language_name'].apply(set).to_dict()
fl_spec_set      = fl_specs_df.groupby('freelancer_id')['speciality_name'].apply(set).to_dict()
has_portfolio    = set(portfolio_df['freelancer_id'].unique()) if len(portfolio_df) else set()
work_exp_cnt     = work_exp_df.groupby('freelancer_id').size().to_dict() if len(work_exp_df) else {}
fl_domain_map    = freelancers_df.set_index('freelancer_id')['domain'].to_dict()
fl_rate_map      = freelancers_df.set_index('freelancer_id')['estimated_rate'].to_dict()
fl_rate_time_map = freelancers_df.set_index('freelancer_id')['rate_time'].to_dict()
fl_currency_map  = freelancers_df.set_index('freelancer_id')['rate_currency'].to_dict()
fl_projects      = freelancers_df.set_index('freelancer_id')['total_jobs'].to_dict()
fl_bio_vec_map   = freelancers_df.set_index('freelancer_id')['bio_vec'].to_dict()

def _most_recent_work(fid: str):
    rows = work_exp_df[work_exp_df['freelancer_id'] == fid]
    if rows.empty:
        return None
    dates = []
    for _, r in rows.iterrows():
        if r['is_current']:
            dates.append(today)
        elif r.get('end_date'):
            try:
                dates.append(date.fromisoformat(r['end_date']))
            except Exception:
                pass
        elif r.get('start_date'):
            try:
                dates.append(date.fromisoformat(r['start_date']))
            except Exception:
                pass
    return max(dates) if dates else None

fl_recent_work_map = {fid: _most_recent_work(fid) for fid in fl_domain_map.keys()}

# Job lookups
job_req_skills  = job_role_skills_df[job_role_skills_df['is_required']].groupby('job_post_id')['skill_name'].apply(set).to_dict()
job_pref_skills = job_role_skills_df[~job_role_skills_df['is_required']].groupby('job_post_id')['skill_name'].apply(set).to_dict()
job_exp_map     = job_posts_df.set_index('job_post_id')['experience_level'].to_dict()
job_domain_map  = job_posts_df.set_index('job_post_id')['domain'].to_dict()
job_budget_map  = job_roles_df.groupby('job_post_id')['role_budget'].max().to_dict()
job_title_map   = job_posts_df.set_index('job_post_id')['job_title'].to_dict()
job_desc_map    = job_posts_df.set_index('job_post_id')['job_description'].to_dict()
job_desc_vec_map = job_posts_df.set_index('job_post_id')['desc_vec'].to_dict()

EXP_ORD = {'entry': 1, 'intermediate': 2, 'expert': 3}
RATE_TO_MONTHLY = {'hourly': 160.0, 'daily': 20.0, 'weekly': 4.0, 'monthly': 1.0, 'annually': 1.0/12.0}
RECENCY_HALF_LIFE_MONTHS = 24.0
RECENCY_LAMBDA = math.log(2.0) / RECENCY_HALF_LIFE_MONTHS

def rate_to_usd(amount, currency):
    factor = APPROX_CURRENCY_TO_USD.get(currency.upper(), 1.0)
    return amount / factor if factor > 0 else amount

def infer_exp(total_jobs):
    if total_jobs >= 10: return 3
    if total_jobs >= 3:  return 2
    return 1

freelancers_df['inferred_exp'] = freelancers_df['total_jobs'].apply(infer_exp)
fl_exp_map = freelancers_df.set_index('freelancer_id')['inferred_exp'].to_dict()


def compute_features(fid, jp_id):
    """Compute 18 features for a (freelancer, job) pair.
    Must match ml_ranker.py _compute_job_features()."""
    f_sk    = fl_skill_set.get(fid, set())
    f_specs = {sp.lower() for sp in fl_spec_set.get(fid, set())}
    f_langs = fl_lang_set.get(fid, set())
    f_skill_prof_map = fl_skill_prof.get(fid, {})

    j_req = job_req_skills.get(jp_id, set())
    j_prf = job_pref_skills.get(jp_id, set())

    matched_skills = f_sk & j_req
    req_match = len(matched_skills)
    req_total = len(j_req)
    skill_pct = req_match / req_total if req_total else 0.0
    pref_pct  = len(f_sk & j_prf) / len(j_prf) if j_prf else 0.0

    if req_total > 0:
        depth_sum = sum(
            PROFICIENCY_WEIGHT.get((f_skill_prof_map.get(sk) or '').lower(), 0.5)
            for sk in matched_skills
        )
        skill_depth = depth_sum / req_total
    else:
        skill_depth = 0.0

    f_exp = fl_exp_map.get(fid, 1)
    j_exp = EXP_ORD.get(job_exp_map.get(jp_id, 'entry'), 1)
    exp_match = int(f_exp >= j_exp)
    exp_delta = int(max(-2, min(2, f_exp - j_exp)))

    f_rate       = fl_rate_map.get(fid, 30.0)
    f_rate_time  = fl_rate_time_map.get(fid, 'hourly')
    f_currency   = fl_currency_map.get(fid, 'USD')
    monthly_mult = RATE_TO_MONTHLY.get(f_rate_time, 160.0)
    f_monthly_usd = rate_to_usd(f_rate * monthly_mult, f_currency)

    j_budget   = job_budget_map.get(jp_id, 1000.0)
    rate_ok    = int(f_monthly_usd <= j_budget * 1.1)
    rate_ratio = float(min(f_monthly_usd / j_budget, 3.0)) if j_budget > 0 else 1.5

    common_langs = {'English', 'Indonesian'}
    lang_match   = int(bool(f_langs & common_langs))

    j_title  = job_title_map.get(jp_id, '').lower()
    j_desc   = job_desc_map.get(jp_id, '').lower()
    job_text = j_title + ' ' + j_desc[:300]

    speciality_match = int(
        bool(f_specs) and any(
            any(word in j_title for word in sp.split() if len(word) > 3)
            for sp in f_specs
        )
    )
    domain_match = int(
        bool(f_specs) and any(
            any(word in job_text for word in sp.split() if len(word) > 3)
            for sp in f_specs
        )
    )

    # cosine_sim — bio-embedding × job-description-embedding (production proxy:
    # production uses freelancer source-text × job source-text from pgvector;
    # in the synthetic notebook the bio is the closest analogue we have).
    bio_vec = fl_bio_vec_map.get(fid)
    job_vec = job_desc_vec_map.get(jp_id)
    cosine_sim = cosine(bio_vec, job_vec) if (bio_vec is not None and job_vec is not None) else 0.0
    cosine_sim = float(np.clip(cosine_sim, 0.0, 1.0))

    # recency_score — 0..1, exp(-λ × months_since_most_recent_work)
    most_recent = fl_recent_work_map.get(fid)
    if most_recent is None:
        recency_score = 0.0
    else:
        months = max(0, (today - most_recent).days) / 30.0
        recency_score = float(math.exp(-RECENCY_LAMBDA * months))

    # portfolio_relevance — max recency-weighted cosine over freelancer's MANUAL
    # portfolio embeddings. (In production, also unioned with contract_embedding,
    # which is more trustworthy because it carries the client rating + review.
    # We don't have synthetic contract embeddings in 01, so we approximate that
    # signal by sampling the auto-generated portfolio rows too — which mirror
    # what contract_embedding holds in production.)
    portfolio_relevance = 0.0
    for vec, age_months in fl_manual_port_vecs.get(fid, []):
        sim = cosine(vec, job_vec) if job_vec is not None else 0.0
        weighted = sim * math.exp(-RECENCY_LAMBDA * age_months)
        if weighted > portfolio_relevance:
            portfolio_relevance = weighted
    portfolio_relevance = float(np.clip(portfolio_relevance, 0.0, 1.0))

    return {
        'cosine_sim':             cosine_sim,
        'skill_overlap_pct':      float(skill_pct),
        'skill_required_matched': float(req_match),
        'skill_required_total':   float(req_total),
        'skill_preferred_pct':    float(pref_pct),
        'experience_level_match': float(exp_match),
        'exp_delta':              float(exp_delta),
        'rate_in_budget':         float(rate_ok),
        'rate_ratio':             float(rate_ratio),
        'language_match':         float(lang_match),
        'speciality_match':       float(speciality_match),
        'domain_match':           float(domain_match),
        'has_portfolio':          float(1 if fid in has_portfolio else 0),
        'work_exp_count':         float(work_exp_cnt.get(fid, 0)),
        'total_jobs':             float(fl_projects.get(fid, 0)),
        'skill_depth':            float(skill_depth),
        'recency_score':          float(recency_score),
        'portfolio_relevance':    float(portfolio_relevance),
    }


def compute_label(feat):
    """Probabilistic fit label with non-linear interactions and unobserved noise.

    Three real-world dynamics that linear models cannot represent:
      • Multiplicative interactions (skill × exp, skill_depth × √recency, cosine × skill).
      • Non-monotonic over-qualification penalty (exp_delta = +2 → 0.85x multiplier).
      • Unobserved noise — caps achievable AUC ~0.92.

    Cold-start fairness:
      • The √portfolio_relevance term is ONLY added when the freelancer has past
        evidence (work_exp > 0 OR has_portfolio = 1). Otherwise we don't penalise
        absence — we just trust skills + cosine_sim + the rest.
    """
    skill_fit   = feat['skill_overlap_pct']
    skill_depth = feat['skill_depth']
    exp_match   = feat['experience_level_match']
    exp_delta   = feat['exp_delta']
    cosine_sim  = feat['cosine_sim']
    recency     = feat['recency_score']
    port_rel    = feat['portfolio_relevance']
    has_port    = feat['has_portfolio']
    work_exp    = feat['work_exp_count']
    domain_fit  = max(feat['domain_match'],
                      cosine_sim / 0.60 if cosine_sim < 0.60 else 1.0)
    profile_depth = min((work_exp + feat['total_jobs']) / 10.0, 1.0)

    # Interactions
    skill_x_exp = skill_fit * (0.5 + 0.5 * exp_match)
    cos_x_skill = cosine_sim * skill_fit
    depth_x_recency = skill_depth * math.sqrt(max(recency, 0.05))

    # Cold-start gate on portfolio_relevance: only contribute if freelancer has any
    # past-work evidence at all. This keeps beginners on a level playing field.
    has_past_work = (work_exp > 0) or (has_port > 0)
    portfolio_term = math.sqrt(max(port_rel, 0.0)) if has_past_work else 0.0

    # Over-qualification (non-monotonic in exp_delta)
    if exp_delta >= 2:
        over_mult = 0.85
    elif exp_delta < 0:
        over_mult = 0.7
    else:
        over_mult = 1.0

    fit_score = (
        0.22 * skill_x_exp
      + 0.18 * depth_x_recency
      + 0.14 * cos_x_skill
      + 0.10 * portfolio_term
      + 0.08 * domain_fit
      + 0.08 * cosine_sim
      + 0.06 * feat['rate_in_budget']
      + 0.06 * feat['language_match']
      + 0.04 * feat['speciality_match']
      + 0.04 * profile_depth
    ) * over_mult

    # Unobserved noise — caps achievable AUC ~0.92
    fit_score = fit_score + float(np.random.normal(0.0, 0.04))

    log_odds = 9.0 * (fit_score - 0.50)
    prob = 1.0 / (1.0 + np.exp(-log_odds))
    return int(np.random.random() < prob)


# Sanity-check
sample_fid  = freelancers_df.iloc[0]['freelancer_id']
sample_jp   = job_posts_df.iloc[0]['job_post_id']
sample_feat = compute_features(sample_fid, sample_jp)
print(f'Feature helpers ready. Features: {len(sample_feat)} (target: 18)')
for k, v in sample_feat.items():
    print(f'  {k:<24s}: {v}')
````

````python
# ── ML training pairs — fit-based label ──────────────────────────────────────
#
# Strategy: sample same-domain pairs (likely positives) and cross-domain
# pairs (likely negatives) for each freelancer. Label by fit, not by outcome.
#
# This means every freelancer is treated equally regardless of their history.
# A new freelancer with good skills gets label=1 for matching jobs.

all_fids    = list(freelancers_df['freelancer_id'])
all_jp_ids  = list(job_posts_df['job_post_id'])

# Domain → job_post_id lookup for stratified sampling
domain_to_jobs = {}
for dom in ALL_DOMAINS:
    domain_to_jobs[dom] = list(job_posts_df[job_posts_df['domain'] == dom]['job_post_id'])

ml_rows    = []
seen_pairs = set()

for fid in all_fids:
    f_dom = fl_domain_map[fid]

    # Same-domain: high chance of positive label
    same_pool = domain_to_jobs.get(f_dom, [])
    n_same    = min(8, len(same_pool))
    same_sample = random.sample(same_pool, n_same) if n_same > 0 else []

    # Cross-domain: predominantly negative label
    cross_doms = [d for d in ALL_DOMAINS if d != f_dom]
    cross_pool = [jp for d in cross_doms for jp in domain_to_jobs.get(d, [])]
    n_cross    = min(5, len(cross_pool))
    cross_sample = random.sample(cross_pool, n_cross) if n_cross > 0 else []

    for jp_id in same_sample + cross_sample:
        if (fid, jp_id) not in seen_pairs:
            seen_pairs.add((fid, jp_id))
            feat  = compute_features(fid, jp_id)
            label = compute_label(feat)
            feat.update({'freelancer_id': fid, 'job_post_id': jp_id, 'label': label})
            ml_rows.append(feat)

ml_df = pd.DataFrame(ml_rows)

pos_rate = ml_df['label'].mean()
print(f'ML training pairs:  {len(ml_df):,}')
print(f'Positive (fit=1):   {ml_df["label"].sum():,}  ({pos_rate:.1%})')
print(f'Negative (fit=0):   {(ml_df["label"]==0).sum():,}  ({1-pos_rate:.1%})')
print(f'\nLabel breakdown by experience level:')
for f_exp_label, f_exp_val in [('entry (0-2 jobs)', [1]), ('intermediate (3-9)', [2]), ('expert (10+)', [3])]:
    mask = ml_df['total_jobs'].apply(infer_exp).isin(f_exp_val)
    sub  = ml_df[mask]
    print(f'  {f_exp_label:22s}: {len(sub):5,} pairs | pos={sub["label"].mean():.1%}')
````

````python
# ── Save all CSVs ─────────────────────────────────────────────────────────────
# Strip embedding-vector columns before writing to CSV — they're list-of-floats
# and don't serialize cleanly. Embeddings live in embeddings_cache.parquet
# (already saved by save_embedding_cache()). The training pairs CSV holds
# the 19 numeric features that 02_train_model.ipynb actually consumes.

freelancers_save_df = freelancers_df.drop(columns=['bio_vec'], errors='ignore')
job_posts_save_df   = job_posts_df.drop(columns=['desc_vec'], errors='ignore')

to_save = {
    'skills':                  skills_df,
    'specialities':            specialities_df,
    'languages':               languages_df,
    'freelancers':             freelancers_save_df,
    'freelancer_skills':       fl_skills_df,
    'freelancer_specialities': fl_specs_df,
    'freelancer_languages':    fl_langs_df,
    'work_experiences':        work_exp_df,
    'educations':              education_df,
    'portfolios':              portfolio_df,
    'job_posts':               job_posts_save_df,
    'job_roles':               job_roles_df,
    'job_role_skills':         job_role_skills_df,
    'ml_training_pairs':       ml_df,
}

print('=== SAVING DATA ===')
for name, df in to_save.items():
    path = f'{DATA_DIR}/{name}.csv'
    df.to_csv(path, index=False)
    print(f'  {name:30s}  {len(df):6,d} rows  →  {path}')

print('\nAll data saved!')
print(f'\nTraining pairs summary:')
feature_cols = [c for c in ml_df.columns if c not in ['freelancer_id','job_post_id','label']]
print(f'  Total:    {len(ml_df):,}')
print(f'  Features: {len(feature_cols)} (target: 18)')
print(f'  Pos rate: {ml_df["label"].mean():.1%}  (target: ~30% for probabilistic fit label)')
print(f'  NaN cols: {ml_df.isnull().sum().sum()} (expected: 0)')
print(f'\nFeature names:')
for c in feature_cols:
    print(f'  - {c}')
````
